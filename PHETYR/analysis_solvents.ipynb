## Analysis of solvents

```python
%%bash

for sol in "che" "ethoh"  "metoh" "dmso"
do
    edr="../data/${sol}_n1000_mini.edr"
    if [ ! -f $edr ]; then
        edr="../data/${sol}_n500_mini.edr"
    fi
    out="../analysis/${sol}_n500_mini_pot.xvg"
    gmx energy -f $edr -o $out <<EOF
10
EOF
done

for sol in "tol" "ben" "hexanol" "octanol"
do
    edr="../data/${sol}_n500_mini.edr"
    out="../analysis/${sol}_n500_mini_pot.xvg"
    gmx energy -f $edr -o $out <<EOF
11
EOF
done

sol="acetone"
edr="../data/${sol}_n1000_mini.edr"
out="../analysis/${sol}_n1000_mini_pot.xvg"
gmx energy -f $edr -o $out <<EOF
12
EOF

xtc="../data/h2o_box_mini.edr"
out="../analysis/h2o_box_mini_pot.xvg"
gmx energy -f $xtc -o $out <<EOF
7
EOF
```

```python
solvents = ["che", "ben", "tol", "octanol", "hexanol",
            "ethoh", "metoh", "dmso", "acetone"]
```

```python
fig, ax = plt.subplots(5,2,figsize=(10,10))
ax = ax.flatten()
for i, sol in enumerate(solvents):
    filename = "../analysis/%s_n1000_mini_pot.xvg"%sol
    try: 
        data = np.loadtxt(filename, comments=["#", "@"])
    except FileNotFoundError as e:
        print (e)
        try:
            filename = "../analysis/%s_n500_mini_pot.xvg"%sol
            data = np.loadtxt(filename, comments=["#", "@"])
        except FileNotFoundError as e:
            print (e)
            break
    ax[i].plot(data[:,0], data[:,1], label='%s'%sol)
    ax[i].legend(fontsize=12, borderpad=0.2, handletextpad=0, handlelength=.5)

filename = "../analysis/h2o_box_mini_pot.xvg"
data = np.loadtxt(filename, comments=["#", "@"])
ax[-1].plot(data[:,0], data[:,1], label='H$_2$O')
ax[-1].legend(fontsize=12, borderpad=0.2, handletextpad=0, handlelength=.5)

#ax[-1].set_ylim(-135000,-0)
[ax[i].set_ylabel('V (kJ/mol)') for i in range(0,10,2)]
ax[-1].set_xlabel('Steps'); ax[-2].set_xlabel('Steps')
plt.tight_layout(w_pad=0, h_pad=0)
```
